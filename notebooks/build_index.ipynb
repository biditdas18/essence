```python
import sys
import os

project_root = os.path.abspath("..")
sys.path.append(project_root)

import json
import numpy as np
import faiss
from utils.embedding import embed_text
```

```python
with open('../data/songs.json', 'r') as f:
    songs = json.load(f)
```

```python
song_vectors = []
for song in songs:
    vec = embed_text(song['emotion_summary'])
    song_vectors.append(vec)
```

```python
song_vectors = np.array(song_vectors, dtype=np.float32)
print("Shape:", song_vectors.shape)  # Should be (15, 384)
```

```python
d = song_vectors.shape[1]  # 384
index = faiss.IndexFlatL2(d)
index.add(song_vectors)
print(f"Total vectors indexed: {index.ntotal}")
```

```python
faiss.write_index(index, '../embeddings/songs.index')
```

```python
with open('../embeddings/song_metadata.json', 'w') as f:
    json.dump(songs, f, indent=4)
```

```python
user_input = "I feel like drifting through space, calm but weightless"
query_vec = embed_text(user_input).reshape(1, -1)

D, I = index.search(query_vec, k=3)  # top 3 matches
for i in I[0]:
    print(f"🎵 {songs[i]['title']} – {songs[i]['artist']}")
    print(f"    {songs[i]['emotion_summary']}")
```
